## Generating tags/features from free text available

```python
def extract_component(text):
    text = text.lower()
    if "wiring harness" in text:
        return "wiring_harness"
    elif "horn switch" in text:
        return "horn_switch"
    elif "spoke cover" in text:
        return "spoke_cover"
    elif "module" in text:
        return "steering_wheel_module"
    elif "steering wheel" in text:
        return "steering_wheel"
    else:
        return "unknown"

def extract_action(text):
    text = text.lower()
    if "replacement" in text:
        return "replacement"
    elif "repair" in text:
        return "repair"
    else:
        return "unknown"

df["component_tag"] = df["global_labor_code_description"].apply(extract_component)
df["action_tag"] = df["global_labor_code_description"].apply(extract_action)
df.head()
```

```python
# Taking Columns with object values -> String values
str_cols = df.select_dtypes(include=object).columns
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())
```

```python
# Handling Null values
df = df.drop(['campaign_nbr'],axis=1)
df.isnull().mean().sort_values(ascending=False)
for col in str_cols:
    df[col] = df[col].fillna("Unknown")
float_cols = df.select_dtypes(include="float").columns
df[float_cols] = df[float_cols].fillna(0.0) #
```

## Identifying Critical Columns

### Component vs Failure Count

```python
df["component_tag"].value_counts().plot(kind="bar")
plt.title("Failure Count by Component")
plt.xlabel("Component")
plt.ylabel("Number of Failures")
plt.tight_layout()
plt.show()
```

### Component vs Total Cost

```python
df.groupby("component_tag")["totalcost"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Total Cost by Component")
plt.xlabel("Component")
plt.ylabel("Total Cost")
plt.tight_layout()
plt.show()
```

### Plant vs Failure Count

```python
df["plant"].value_counts().plot(kind="bar")
plt.title("Failure Count by Plant")
plt.xlabel("Plant")
plt.ylabel("Number of Failures")
plt.tight_layout()
plt.show()
```

# Actionable Recommendations for Stakeholders

#### Based on the analysis, the following recommendations are proposed:

- Prioritize High-Impact Components
Components that show both high replacement frequency and high total cost should be prioritized for design review, supplier evaluation, or enhanced quality checks.

- Investigate Replacement-Only Failure Patterns
The absence of repair actions suggests limited opportunities for partial fixes. Stakeholders should evaluate whether design improvements or preventive maintenance could reduce the need for full replacements.

- Plant-Level Process Review
Variations in failure counts across plants indicate potential differences in process quality, tooling, or operating conditions. Targeted audits and best-practice sharing between plants could help reduce overall failure rates.

- Improve Data Capture Practices
Missing numeric values and inconsistent descriptions highlight opportunities to improve upstream data collection, which would enable more precise cost and effort analysis in the future.

## Rename for Semantic Alignment

```python
df1 = df.rename(columns={
    "global_labor_code_description": "labor_description"
})

df2 = df2.rename(columns={
    "failure_condition_-_failure_component": "labor_description"
})
```

```python
merged_df = df1.merge(
    df2,
    on="labor_description",
    how="left",
    suffixes=("_task1", "_task2")
)
print(f'{merged_df.shape}')
merged_df.head()
```

```python
merged_df.info()
```

### While integrating the datasets, it was observed that several fields from the secondary dataset remained null after the merge. This indicates that the second dataset serves as a higher-level reference or semantic layer rather than a one-to-one transactional mapping. A left join was intentionally used to preserve all repair events from the primary dataset, even when supplementary metadata was unavailable
